#Certificate logic function

```python
def certificate_generator(name, courseName, courseLevel, date):

  #how to store
  pdf_file_name = "/content/drive/MyDrive/CLASS ACTIVITY/certificates/" + name + "-" + courseName + "-" + courseLevel + ".pdf"

  #create canvas
  c = canvas.Canvas(pdf_file_name, pagesize=landscape(A4))

  # write all these info into cert template
  c.drawImage("/content/drive/MyDrive/CLASS ACTIVITY/certificate_template.jpg", 0, 0, width=A4[1], height=A4[0])

  center_x = c._pagesize[0] / 2
  center_y = c._pagesize[1] / 2

  # 1. name
  c.setFont("Lora-Bold", 30)
  c.drawCentredString(center_x, center_y - 46, name)

  # 2. course name
  c.setFont("Lora-Bold", 28)
  c.drawCentredString(center_x, center_y - 105, courseName + "-" + courseLevel)

  # 3. date
  c.setFont("Lora-Bold", 28)
  c.drawCentredString(center_x + 190, center_y - 160, date)

  # 4. cert ID
  cert_id = "Cert ID: " + str(int(pd.Timestamp.now().timestamp())) #generate id based on timestamp
  c.setFont("Lora-Regular", 12)
  c.drawCentredString(center_x + 266, center_y - 230, cert_id.upper())


  c.save()
```

#Do the cert generator

##Some function
1. .iterrows() -> give loopable like array

```python
for i, x in df.iterrows():
  certificate_generator(x["Name"], x["Course"], x["CourseLevel"], x["Date"])

  print("The certificate are successfully generated for " + x["Name"])


#print(str(len(df)) + "The certificate are successfully generated.")
```
